```r
# Problem 4
niterations = 1e5 # number of iterations
below = rep(0, niterations) # set storage
set.seed(2000)

for(i in 1:niterations)
{
    r = rnorm(45, mean = 0.05/253, sd = 0.23 / sqrt(253)) # generate random numbers
    logPrice = log(1e6) + cumsum(r)
    minLogPrice = min(logPrice) # minimum price over 45 days
    below[i] = as.numeric(minLogPrice < log(950000))
}
mean(below)
```

```r
# Problem 5
niterations = 1e5
profit = rep(0, niterations)
n_days_simulation = 100
mean = 0.23/253
stddev = 0.05/sqrt(253)
initial_value = 1e6
profit_target = initial_value + 1e5

for(i in 1:niterations)
{
    r = rnorm(n_days_simulation, mean = mean, sd = stddev)
    logPrice = log(initial_value) + cumsum(r)
    profit[i] = as.numeric(any(logPrice > log(profit_target)))
}
mean(profit)
```

```r
# Problem 6
niterations = 1e5
loss = rep(0, niterations)
n_days_simulation = 100
mean = 0.05/253
stddev = 0.23/sqrt(253)
initial_value = 1e6
stop_loss_target = initial_value - 50000

for(i in 1:niterations)
{
    r = rnorm(n_days_simulation, mean = mean, sd = stddev)
    logPrice = log(initial_value) + cumsum(r)
    loss[i] = as.numeric(any(logPrice < log(stop_loss_target)))
}
mean(loss)
```

```r
# Problems 7 and 8
niterations = 1e5
expected_profit = rep(0, niterations)
expected_return = rep(0, niterations)
n_days_simulation = 100
mean = 0.05/253
stddev = 0.23/sqrt(253)
initial_invesment = 50000
initial_value = 1e6
profit = 1e5
profit_target = initial_value + profit
loss = -50000
stop_loss_target = initial_value + loss

for(i in 1:niterations)
{
    r = rnorm(n_days_simulation, mean = mean, sd = stddev)
    logPrice = log(initial_value) + cumsum(r)
    
    profit_target_triggered = any(logPrice >= log(profit_target))
    stop_loss_target_triggered = any(logPrice <= log(stop_loss_target))
    
    if(profit_target_triggered & stop_loss_target_triggered)
    {
        days_to_profit_target_trigger = which(logPrice >= log(profit_target))[1]
        days_to_stop_loss_target_trigger = which(logPrice <= log(stop_loss_target))[1]
        
        if(days_to_profit_target_trigger < days_to_stop_loss_target_trigger)
        {
            expected_profit[i] = profit
        }
        else
        {
            expected_profit[i] = loss
        }
        
        total_days_to_trade = min(days_to_profit_target_trigger, days_to_stop_loss_target_trigger)
     }
    
    if(profit_target_triggered & !stop_loss_target_triggered)
    {
        expected_profit[i] = profit
        total_days_to_trade = days_to_profit_target_trigger
    }
    
    if(!profit_target_triggered & stop_loss_target_triggered)
    {
        expected_profit[i] = loss
        total_days_to_trade = days_to_stop_loss_target_trigger
    }
    
    if(!profit_target_triggered & !stop_loss_target_triggered)
    {
        expected_profit[i] = exp(logPrice[n_days_simulation]) - initial_value
        total_days_to_trade = n_days_simulation
    }
    
    expected_return[i] = (expected_profit[i]/initial_invesment - 1)/total_days_to_trade
}
mean(expected_profit)
mean(expected_return)
```

```r
# Simulating a geometric random walk
set.seed(2000)
niterations = 9
n = 253
mean = 0.05/n
stddev = 0.23/sqrt(n)
par(mfrow=c(3, 3))

for(i in 1:niterations)
{
    logr = rnorm(n, mean = mean, sd = stddev)
    price = c(120, 120 * exp(cumsum(logr)))
    plot(price, type = "b")
}
```
